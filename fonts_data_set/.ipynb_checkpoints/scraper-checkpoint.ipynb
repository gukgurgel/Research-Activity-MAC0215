```python
from bs4 import BeautifulSoup as bs
import requests
from zipfile import ZipFile
import os
```

```python
DOMAIN = "https://www.dafont.com/mtheme.php?id=5&page="
n_pages = 5 # each page has 200 fonts
URLs = [DOMAIN + str(i + 1) + "&fpp=200" for i in range(n_pages)]
DOWNLOAD = '\xa0Download\xa0'
```

```python
URLs
```

```python
def get_soup(url):
    user_agent = 'Mozilla/5.0'
    headers = {'User-Agent': user_agent}
    return bs(requests.get(url, headers = headers).text, "html.parser")
```

```python
fonts = ['']*(n_pages*200)
```

```python
counter = 0
for URL in URLs:
    for link in get_soup(URL).find_all('a'):
        download_link = link.get('href')
        contents = link.contents
        if DOWNLOAD in contents:
            print(download_link)
            font_name = download_link[22:]
            with open(font_name, "wb") as file:
                response = requests.get("https:" + download_link)
                file.write(response.content)
                fonts[counter] = font_name
                counter += 1
```

```python
with open('fonts_list.txt', 'w') as file:
    file.write(str(fonts))
```

# Extracting Files

```python
for font in fonts:

    # opening the zip file in READ mode 
    with ZipFile(font, 'r') as zip: 
        os.chdir('fonts')
        os.mkdir(font)
        zip.extractall(path = font)
        os.chdir('..') #leave fonts dir to base dir
```
